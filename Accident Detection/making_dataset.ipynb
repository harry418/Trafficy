# Generate new dataset

**Checking files distribution in given directory.**

```
import os

  # dir is your directory path
num_files = os.listdir('/content/accidents')
print(num_files)
```

```
for i in num_files[:-1]:
  path, dirs, files = next(os.walk('/content/accidents'+'/'+i))
  file_count = len(files)
  print('{} contain {} number of files.'.format(i,str(file_count)))
```

**Spiltting into train, test and validation in 0.7,0.2.0.1 respectively**

```
!pip install split-folders
import splitfolders
```

```
print('spiltting dataset in train, test and validation')
splitfolders.ratio("/content/accidents", output="/content/crashs", seed=1337, ratio=(.1, .7, .2), group_prefix=None) # default values
```

**Check file diatribution in train, test and validation folder**

```
check = ['train','test','val']
for fol in check:
    for i in num_files[:-1]:
        path, dirs, files = next(os.walk('/content/crashs/'+fol+'/'+i))
        file_count = len(files)
        print('{}  : {} ===>> {} files.'.format(fol,i,str(file_count)))
```
